### Tensor Reducing operations

an operation that decreases the number of elements in a tensor by summarizing them across one or more dimensions.

```python
import numpy as np
import torch
```

```python
vector = torch.tensor([10, 20, 30, 40, 50, 60], dtype=torch.float32)
```

```python
print(f"max = {vector.max()}")
print(f"min = {vector.min()}")
print(f"mean = {vector.mean()}")
print(f"sum = {vector.sum()}")
print(f"product = {vector.prod()}")
print(f"SD = {vector.std()}")
```

### Create rank2 tensor

```python
matrix = torch.tensor([[100, 20], [30, 40], [50, 60]], dtype=torch.float32)
matrix
```

```python
print(f"max = {matrix.max()}")
print(f"min = {matrix.min()}")
print(f"mean = {matrix.mean()}")
print(f"sum = {matrix.sum()}")
print(f"product = {matrix.prod()}")
print(f"sd = {matrix.std()}")
```

### Perform operations column wise

```python
print(f"max = {matrix.max(dim=0).values}")
print(f"min = {matrix.min(dim=0).values}")
print(f"mean = {matrix.mean(dim=0)}")
print(f"sum of all values = {matrix.sum(dim=0)}")
```

### Perform operations Row wise

```python
print(f"max = {matrix.max(dim=1).values}")
print(f"min = {matrix.min(dim=1).values}")
print(f"mean = {matrix.mean(dim=1)}")
print(f"sum = {matrix.sum(dim=1)}")
```
