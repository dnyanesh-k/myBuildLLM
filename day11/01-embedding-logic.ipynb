### embedding logic using lookup table

```python
import numpy as np
import torch
```

```python
vocab_size = 10

dimensions = 5
```

```python
lookup_table = torch.rand(vocab_size, dimensions, requires_grad=True)
lookup_table
```

```python
lookup_table.shape
```

```python
token_ids = torch.tensor([1, 6, 8, 9])
```

```python
embeddings = lookup_table[token_ids]
embeddings
```

### Use tiktoken to generate embeddings

```python
import tiktoken
tokenizer = tiktoken.get_encoding('gpt2')
tokenizer.n_vocab
```

```python
# create a lookup table
lookup_table = torch.randn(tokenizer.n_vocab, 5, requires_grad=True)
lookup_table
```

```python
lookup_table.shape
```

```python
# input 
sentence = "I love machine learning"


# get the tokens for the sentence
token_ids = tokenizer.encode(sentence)
token_ids
```

```python
# get the embeddings for token_ids
embeddings = lookup_table[token_ids]
embeddings
```
