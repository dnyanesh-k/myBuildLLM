# Tiktoken: 
It is a fast `BPE (Byte Pair Encoding)` tokenizer developed by `OpenAI`. It converts **raw text** into **integers (tokens)** that models like GPT-4o can process. It’s written in **Rust** for performance, making it significantly **faster than standard Python-based BPE implementations**.

## Why use it?
- **Performance:** It handles large-scale text processing much faster
- **Cost & Context Management:** It allows you to calculate exactly how many tokens a string contains, which is critical for staying within context window limits and estimating API costs.
- **Model Parity:** It ensures your local preprocessing exactly matches how OpenAI’s models see the text (e.g., using o200k_base for GPT-4o).

## Limitations
- **OpenAI Specific:** It is purpose-built for OpenAI’s specific encodings. While it supports BPE, it isn't a general-purpose library for training new tokenizers from scratch like Hugging Face’s tokenizers.
- **Loss of Formatting:** Like all BPE tokenizers, it can struggle with whitespace-heavy code or specialized mathematical notation if those patterns weren't frequent in the training data.
- **No "In-Library" Training:** You generally use it to load existing encodings; it lacks the high-level API to easily train and save custom tokenizers for non-OpenAI models.


```python
import tiktoken
```

```python
# create a tokenizer using a model
tokenizer = tiktoken.get_encoding('gpt2')
```

```python
corpus = "I love machine learning, welcome to build my llm"
```

```python
# encode the corpus to get the tokens
tokens = tokenizer.encode(corpus)
tokens
```

```python
# get the original corpus using the tokens
original_text = tokenizer.decode(tokens)
original_text
```

```python
print(f"Size of tiktoken vocabulory = {tokenizer.n_vocab}")
```

```python
for token in tokens:
    print(f'{token} = {tokenizer.decode([token])}')
```

```python
# create tokenizer using gpt-4o
tokenizer_4o = tiktoken.get_encoding('o200k_base')
```

```python
print(tokenizer_4o.n_vocab)
```

```python
tokenizer_4o.encode(corpus)
```
