### Tensor GPU support

```python
import numpy as np
import torch
```

```python
cpu_tensor = torch.tensor([10, 20, 30, 40, 50])
cpu_tensor.device
```

Create tensor on GPU

```python
# check if cuda is available
torch.cuda.is_available()
```

```python
# check if mps is available
torch.backends.mps.is_available()
```

```python
device_name = 'cpu'
if torch.cuda.is_available():
    device_name = torch.device('cuda')
elif torch.backends.mps.is_available():
    device_name = torch.device('mps')

print(f"selected device = {device_name}")    
```

```python
# create tensor on gpu
gpu_tensor = torch.tensor([10, 20, 30, 40, 50], device = device_name)
gpu_tensor.device
```

```python
# move tensor to gpu
gpu_tensor =  cpu_tensor.to(device_name)
gpu_tensor.device
```
