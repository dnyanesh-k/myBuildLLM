## word2vector demo using existing model

```python
import gensim.downloader as api
```

```python
# download the model
model = api.load("word2vec-google-news-300")
```

```python
# get embeddings for required words
model["computer"]
```

```python
model.similarity("king", "man")
```

```python
model.similarity("king", "queen")
```

```python
model.similarity("king", "apple")
```

```python
model.similarity("apple", "orange")
```

```python
model.similarity("machine", "ai")
```

```python
model.similarity("boy", "girl")
```

```python
model.similarity("cat", "dog")
```
