Word level tokenizer

```python
import numpy as np
import re
```

```python
class WordTokenizer:
    def __init__(self, corpus):
        # split the corpus into words
        all_words = re.split(r'[,.?"\'?! ]', corpus)

        # get the final list of words
        words = []
        for word in all_words: # Looks at every item in your original list.
            if word.strip(): # Checks if the word has actual characters (ignoring spaces
                words.append(word.lower()) # Converts the valid words to lowercase and puts them into the new list.

        # add the special tokens
        words.extend(['[UNK]', '<|im_start|>', '<|im_end|>']) 

        # build the dictionaries
        # enumerate gives you (index, item)
        self.encoding_dictionary = {word:token_id for token_id, word in enumerate(words)}
        self.decoding_dictionary = {token_id:word for token_id, word in enumerate(words)}     

    def encode(self, text):
        # split the data into words
        words = re.split(r'[,.?"\'?! ]', text)
        return [self.encoding_dictionary.get(word.lower()) if self.encoding_dictionary.get(word.lower()) else self.encoding_dictionary.get('[UNK]') for word in words ]
    
    def decode(self, tokens):
        return [self.decoding_dictionary.get(token) for token in tokens]
    
    def n_vocab(self):
        return len(self.encoding_dictionary.items())
```

```python
# create a vocabulary
corpus = "I love machine learning, i love artificial intelligence. Welcome to build your own llm. Tokenization is fascinating."
# corpus = "I love machine learning. Welcome to build your own llm. Tokenization is facinating."

# create the tokenizer
tokenizer = WordTokenizer(corpus)

# find the vocab size
print(f'Vocab Size = {tokenizer.n_vocab()}')
```

```python
new_text = "Welcome to machine learning"
tokens = tokenizer.encode(new_text)
tokens
```

```python
# decode the tokens back to text
tokenizer.decode(tokens)
```
