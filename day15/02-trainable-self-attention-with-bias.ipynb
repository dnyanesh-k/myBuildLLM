## trainable self attention msm with bias
- In deep learning, a bias (b) is an extra trainable number added to the output of a matrix multiplication (XW).

- The fundamental formula for a layer changing an input `X`using weights `W` and bias `(b)` is: `Output = XW+b`

```python
import numpy as np
import torch.nn as nn
import math
import torch
```

```python
class SelfAttentionV2(nn.Module):

    def __init__(self, dimension_input, dimesion_output, bias = False):
        super().__init__()

        # create the learned weight metrices for Q, K, V
        self.weight_query = torch.nn.Linear(dimension_input, dimesion_output, bias=bias)
        self.weight_key = torch.nn.Linear(dimension_input, dimesion_output, bias=bias)
        self.weight_value = torch.nn.Linear(dimension_input, dimesion_output, bias=bias)
        
    def forward(self, inputs):
        queries = self.weight_query(inputs)
        keys = self.weight_key(inputs)
        values = self.weight_value(inputs)

        attention_scores = queries @ keys.T

        dimesion_keys = inputs.shape[-1]
        scaled_weights = attention_scores / math.sqrt(dimesion_keys)  

        attention_weights = torch.softmax(scaled_weights, dim=-1)

        context_vector = attention_weights @ values
        
        return context_vector 
```

```python
# input embeddings
inputs = torch.tensor([
   [0.43, 0.15, 0.89], # Your 
   [0.55, 0.87, 0.66], # journey
   [0.57, 0.85, 0.64], # starts
   [0.22, 0.58, 0.33], # with
   [0.77, 0.25, 0.10], # one
   [0.05, 0.80, 0.55]  # step
])
```

```python
torch.manual_seed(123)
dimesion_input = inputs.shape[1]
dimension_output = 2
```

```python
self_attention = SelfAttentionV2(dimesion_input, dimension_output, bias=False)
```

```python
context_vector = self_attention(inputs)
context_vector
```
