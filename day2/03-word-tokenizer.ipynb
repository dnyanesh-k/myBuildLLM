Word Level Tokenizer

```python
class WordTokenizer:
    def __init__(self, vocab):
        # persist the vocabulary
        self.vocab = vocab

        # build the dictonaries
        self.stoi = {ch:i for i, ch in enumerate(vocab)}
        self.itos = {i:ch for i, ch in enumerate(vocab)}

    def encode(self, words):
        # convert every character to corresponding integer (token id) value
        return [self.stoi[word] for word in words]

    def decode(self, encoded_value):
        # convert every integer to corresponding character
        return " ".join([self.itos[num] for num in encoded_value])

    def print_info(self):
        print(self.stoi)
        print(self.itos)
```

```python
corpus = "I love machine learning, i love artificial intelligence and I also like coding and I do gym regularly"
```

```python
# creates the list of every word from the corpus
words = corpus.split()

# this form will include the punctuation symbol
words
```

```python
import re

# find the words from the corpus
words =  re.split(r'[,.?|!"\' ]', corpus)

# remove the empty words and empty spaces
words = [word.strip() for word in words if word.strip()]

# find the unique words and sort them
words = sorted(set(list(words)))
words
```

```python
# create the tokenizer
tokenizer = WordTokenizer(words)
tokenizer.print_info()
```

```python
# test the tokenizer
new_text = "I do coding regularly."

new_text_words = list(re.split(r'[,.?|!"\ ]', new_text))
new_text_words = [new_word.strip() for new_word in new_text_words if new_word.strip()]

encoded_value = tokenizer.encode(new_text_words)
encoded_value
```

```python
tokenizer.decode(encoded_value)
```
