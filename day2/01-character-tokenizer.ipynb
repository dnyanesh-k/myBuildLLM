Character Tokenizer
import required packages

```python
import numpy as np
```

get the data

```python
corpus = "I love machine learning. I love Artifical intelligence"
```

tokenization logic

```python
# build a dictionary to hold the token ids of every character in the corpus
stoi = {}

# build the new dictoionary to hold the tokens for every token ids
itos = {}
```

```python
# build the vocabulary (collection of unique characters)
vocabulory = sorted(set(list(corpus)))
vocabulory
```

```python
token_id = 0
for ch in vocabulory:
    print(f"ch = {ch}, Tokem Id = {token_id}")
    token_id +=1
```

```python
for (ch, token_id) in enumerate(vocabulory):
    # get integer representation of every token and add mapping to the dictionary
    stoi[ch] = token_id

    # get chaarcter represenation of every token and add mapping to dictionary
    itos[token_id] = ch    
```

```python
# get integer representation of every token and add it ot mapping dictionary
stoi = {ch: i for i, ch in enumerate(vocabulory)}

# get character represemtation of every token and add it to mapping dictionary
itos = {i:ch for i, ch in enumerate(vocabulory)}
```
