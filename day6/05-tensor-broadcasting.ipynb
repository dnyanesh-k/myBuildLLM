## tensor broadcasting

```python
import numpy as np
import torch
```

```python
vector = torch.tensor([10, 20, 30, 40, 50])
vector
```


## Arithmetic broadcast operations

```python
print(f"vector + 10 = {vector + 10}")
print(f"vector - 10 = {vector - 10}")
print(f"vector / 10 = {vector / 10}")
print(f"vector // 10 = {vector // 10}")
print(f"vector * 10 = {vector * 10}")
print(f"vector ** 2 = {vector ** 2}")
```

## Arithmetic broadcast operations using two tensors

```python
t1 = torch.tensor([10, 20, 30, 40, 50])
t2 = torch.tensor([60, 70, 80, 90, 100])

print(f"t1 + t2 = {t1 + t2}")
print(f"t2 - t1 = {t2 - t1}")
print(f"t1 * t2 = {t1 * t2}")
print(f"t2/t1 = {t2/t1}")
```

### arithmetic broadcast operations using a rank2 tensor

```python
t1 = torch.tensor([[10, 20], [30, 40]])
t2 = torch.tensor([[50, 60], [70, 80]])

print(f"t1 + t2 = {t1 + t2}")
```

### Comparison broadcast operations

```python
print(f"vector > 4 = {vector > 40}")
print(f"vector < 4 = {vector < 40}")
print(f"vector <= 40 = {vector <= 40}")
print(f"vector >= 40 = {vector >= 40}")
print(f"vector == 40 = {vector == 40}")
print(f"vector != 40 = {vector != 40}")
```
