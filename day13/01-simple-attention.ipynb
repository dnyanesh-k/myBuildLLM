## simple attention mechanism

```python
import numpy as np
import torch
import math
```

### get the input

```python
# input string
input_str = "Your journey starts with one step"
```

```python
# get the embeddings of all the tokens
inputs = torch.tensor([
   [0.43, 0.15, 0.89], # Your
   [0.55, 0.87, 0.66], # journey
   [0.57, 0.85, 0.64], # starts
   [0.22, 0.58, 0.33], # with
   [0.77, 0.25, 0.10], # one
   [0.05, 0.80, 0.55]  # step
])
```

```python
# take journey as an example 
query = inputs[1]
query
```

### calculate attention scores

```python
# attentions cores of journey wrt to all the embeddings
attention_scores_journey = torch.empty(inputs.shape[0])

# calculate the attention score with all embeddings
for index, embedding in enumerate(inputs):
    attention_scores_journey[index] = torch.dot(query, embedding)

attention_scores_journey    
```

```python
attention_scores_journey.sum()
```

### calculate attention weights

```python
# get the total attention scores
attention_scores_journey_sum = attention_scores_journey.sum()
attention_weights_journey = attention_scores_journey/attention_scores_journey_sum
attention_weights_journey
```

```python
attention_weights_journey.sum()
```

## apply softmax

```python
weights = [0.1455, 0.2278, 0.2249, 0.1285, 0.1077, 0.1656]

# calculate the e_total
e_total = 0
for value in weights:
    e_total += math.exp(value)

# divide every value in the weights by e_total
normalized_weights = [math.exp(value)/e_total for value in weights]    
normalized_weights
```

```python
def my_softmax(x):
    return torch.exp(x) / torch.exp(x).sum(dim=0)

my_softmax(torch.tensor(weights))
```

```python
normalized_weights = torch.softmax(torch.tensor(weights), dim=0)
normalized_weights
```

```python
normalized_weights.sum()
```
