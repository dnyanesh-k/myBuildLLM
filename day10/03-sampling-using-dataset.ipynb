## data sampling using sliding window with a dataset

```python
import numpy as np
import torch
import tiktoken
from torch.utils.data import Dataset, DataLoader
```

```python
with open(r"data.txt", "r") as file:
    raw_text = file.read()
```

### create a dataset with input and target tokens

```python
class DreamLLMDataset(Dataset):
    def __init__(self, raw_text, context_length = 10, stride = 1):
        # craete the tokenizer
        tokenizer = tiktoken.get_encoding('gpt2')

        # create the tokens
        tokens = tokenizer.encode(raw_text)

        # store the input and target tensors
        self.input_ids = []
        self.target_ids = []

        # create the input and target tensors
        for i in range(1, len(tokens) - context_length, stride):
            input_chunk = tokens[i : i + context_length]
            target_chunk = tokens[i + 1 : i + context_length + 1]

            self.input_ids.append(torch.tensor(input_chunk))
            self.target_ids.append(torch.tensor(target_chunk))

    def __len__(self):
        return len(self.input_ids)        
    
    def __getitem__(self, index):
        return self.input_ids[index], self.target_ids[index]
```

```python
# create the dataset
dataset = DreamLLMDataset(raw_text=raw_text)

# craete a dataloader
data_loader = DataLoader(dataset=dataset, batch_size=5)
```

```python
#  get the iterator
data_loader_iterator = iter(data_loader)
```
