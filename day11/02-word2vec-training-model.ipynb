## word to vector demo

```python
import numpy as np
from gensim.models import Word2Vec
```

```python
# input for training word2vec model
sentences = [
    ["i", "love", "ai"],
    ["i", "love", "machine", "learning"],
    ["ai", "is", "fun"],
    ["machine", "learning", "is", "powerful"],
    ["i", "enjoy", "learning"],
    ["deep", "learning", "uses", "neural", "networks"]
]
```

```python
# create a word2vec model
model = Word2Vec(
    sentences = sentences, # input
    vector_size = 10, # dimension
    epochs=100,
    min_count=1,
    sg=1
)
```

```python
# get the vector of required token
model.wv["i"]
```

```python
model.wv["learning"]
```
