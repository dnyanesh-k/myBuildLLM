## Tensor Basics

```python
import numpy as np
import torch
```

## Tensor creation demo
### Scalar tensor (Rank0)

```python
t1 = torch.tensor(10.0)
t1
```

```python
print(f"no of dimensions = {t1.dim()}")
```

```python
t2 = torch.tensor(10.0, requires_grad=True)
t2
```

## Vector Tensor (Rank1)

```python
# list of numbers 
numbers_list = [10, 20, 30, 40, 50]
print(f"Numbers list : {numbers_list}, Type = {type(numbers_list)}")
```

```python
# numpy array of numbers
numbers_array = np.array([10, 20, 30, 40, 50])
print(f"numbers_array = {numbers_array}, type = {type(numbers_array)}")
```

```python
# vector of numbers
number_tensor = torch.tensor([10, 20, 30, 40, 50])
print(f"numbers_tensor : {number_tensor}, type = {type(number_tensor)}")
print(f"no of dim = {number_tensor.ndim}")
```

## matrix Tensor (Rank2)

```python
matrix_tensor = torch.tensor([[10, 20], [30, 40], [50, 60]])
print(f"matrix tensor = {matrix_tensor}, type = {type(matrix_tensor)}")
print(f"no of dim = {matrix_tensor.ndim}")
```

## Matrix Tensor (Rank3)

```python
matrix_tensor_3d = torch.tensor([[[10, 20], [30, 40]], [[50, 60], [70, 80]]])
print(f"matrix_tensor_3d = {matrix_tensor_3d}, type: {type(matrix_tensor_3d)}")
print(f"no of dim = {matrix_tensor_3d.ndim}")
```
